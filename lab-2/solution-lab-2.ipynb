```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import random
import imageio
import os

from tqdm import tqdm
```

# 1. Implement Julia Map as $f(z, c)$

```python
def julia_map(z, c):
    return z ** 2 + c
```

```python
def calculate_image_evolution(z, c, n_iterations):
    evolution = [z]

    for _ in range(n_iterations):
        evolution.append(julia_map(evolution[-1], c))
    
    return evolution
```

```python
golden = (1 + 5 ** 0.5) / 2
c, z = 1 - golden, 0.8 * np.exp(0.3j)
n_iterations = 10
evolution = calculate_image_evolution(z, c, n_iterations)

evolution_re, evolution_im = list(zip(*((step.real, step.imag) for step in evolution)))
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(evolution_re, evolution_im)
plt.scatter([0], [0], color='r')

plt.xlim(-1, 1)
plt.ylim(-1, 1)

plt.xlabel('Re$(z)$', fontsize=14)
plt.ylabel('Im$(z)$', fontsize=14)

for i in range(len(evolution)):
    plt.annotate(i, (evolution_re[i], evolution_im[i]), fontsize=14,
                 xytext=(evolution_re[i] + 0.01, evolution_im[i] + 0.01))
    
plt.grid()
plt.show()
```

# 3. Construct & Plot Julia Set for $c = 1 - \varphi$

```python
def check_for_convergence(z, c, n_iterations, max_module):
    converges = True
    
    for i in range(1, n_iterations + 1):
        z = julia_map(z, c)
        
        if abs(z) > max_module:
            converges = False
            break
    
    ratio = i / n_iterations
    return ratio, converges
```

```python
def construct_julia_set(x_range, y_range, x_min, y_min, x_max, y_max, c, max_module, n_iterations):
    width, height = x_max - x_min, y_max - y_min
    julia_set = np.zeros((x_range, y_range))

    for i in range(x_range):
        for j in range(y_range):

            z = complex(i / x_range * width + x_min, j / y_range * height + y_min)
            ratio, converges = check_for_convergence(z, c, n_iterations, max_module)
            julia_set[i, j] = 1 - np.sqrt(ratio)
    
    return julia_set
```

```python
x_range, y_range = 500, 500

x_min, x_max = -1.5, 1.5
width = x_max - x_min

y_min, y_max = -1.5, 1.5
height = y_max - y_min
```

```python
c = 1 - golden
max_module = 4
n_iterations = 200
```

```python
julia_set = construct_julia_set(x_range, y_range, x_min, y_min, x_max, y_max, c, max_module, n_iterations)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(julia_set, interpolation='nearest', cmap=cm.inferno_r)

xtick_labels = np.linspace(x_min, x_max, int(width / 0.2))
plt.xticks([(x - x_min) / width * x_range for x in xtick_labels],
           labels=[f'{xtick:.1f}' for xtick in xtick_labels])

ytick_labels = np.linspace(y_min, y_max, int(height / 0.2))
plt.yticks([(y - y_min) / height * y_range for y in ytick_labels], 
           labels=[f'{ytick:.1f}' for ytick in ytick_labels])

plt.xlabel('Re$(z_0)$', fontsize=14)
plt.ylabel('Im$(z_0)$', fontsize=14)

plt.show()
```

# 4. Plot Julia Set for Continuous Sequence of Parameter $c$

### Assuming that $c = 0.7785 e^{i\varphi},\;\varphi \in [0, 2\pi)$

```python
max_module = 4
n_iterations = 300
```

```python
n_params = 360
cc = 0.7885 * np.exp(2 * np.pi * 1j * np.linspace(0, 1, n_params))
```

```python
folder = f'./julia-set-{max_module}-{n_iterations}/'

if not os.path.exists(folder):
    os.mkdir(folder)
```

```python
for i in tqdm(range(len(cc))):
    julia_set = construct_julia_set(x_range, y_range, x_min, y_min, x_max, y_max, 
                                    cc[i], max_module, n_iterations)
    plt.figure(figsize=(10, 10))
    plt.imshow(julia_set, interpolation='nearest', cmap=cm.magma)

    xtick_labels = np.linspace(x_min, x_max, int(width / 0.2))
    plt.xticks([(x - x_min) / width * x_range for x in xtick_labels],
               labels=[f'{xtick:.1f}' for xtick in xtick_labels])

    ytick_labels = np.linspace(y_min, y_max, int(height / 0.2))
    plt.yticks([(y - y_min) / height * y_range for y in ytick_labels], 
               labels=[f'{ytick:.1f}' for ytick in ytick_labels])
    
    plt.xlabel('Re$(z_0)$', fontsize=14)
    plt.ylabel('Im$(z_0)$', fontsize=14)

    plt.savefig(folder + f'julia-set-{i}.jpg', bbox_inches='tight')
    plt.close('all')
```

```python
filenames = list(filename for filename in os.listdir(folder) if filename.endswith('jpg'))
filenames = sorted(filenames, key=lambda x: int(os.path.splitext(x)[0].split('-')[-1]))
```

```python
images = []

for filename in filenames:
    images.append(imageio.imread(folder + filename))
    
imageio.mimsave(folder + 'julia-set-motion.gif', images)
```
