```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

# 1. Implement Logistic Map as $f(x, r)$

```python
def logistic_map(x, r):
    return r * x * (1 - x)
```

```python
r = 1.5
xx = np.linspace(-.25, 1.25, 100)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(xx, logistic_map(xx, 1.))
plt.hlines(0, -1, 2, color='r')
plt.xlim(-0.25, 1.25)

plt.xlabel('$x$', fontsize=14)
plt.ylabel(f'logistic map given $r$ = {r}', fontsize=14)
plt.grid()
plt.show()
```

# 2. Explore Impact of Parameter $r$ on Evolution of $x_n$ and Final State $x_{\infty}$

```python
n_iterations = 50
n_last_iterations = n_iterations
```

```python
xx = np.array([0.25, 0.5, 0.75])
rr = np.array([3.2, 3.5, 3.7])
```

```python
def calculate_system_evolution(x, r, n_iterations, n_last_iterations):
    evolution = [x]

    for _ in range(n_iterations):
        evolution.append(logistic_map(evolution[-1], r))

    evolution = evolution[-n_last_iterations:]
    steps = list(range(n_iterations - n_last_iterations, n_iterations))
    
    return steps, evolution
```

```python
fig, ax = plt.subplots(3, 3, figsize=(14, 10), sharex=True, sharey=True, frameon=False)

for i in range(len(xx)):
    for j in range(len(rr)):
        x, r = xx[i], rr[j]
        steps, evolution = calculate_system_evolution(x, r, n_iterations, n_last_iterations)
        
        ax[i, j].plot(steps, evolution, marker='o', linestyle='dashed', linewidth=0.5)
        ax[i, j].set_title(f'$x_0$ = {x}, $r$ = {r}', fontsize=12)
        
        ax[i, j].grid()
        ax[i, j].set_ylim(0, 1)
#         ax[i, j].set_xlabel('step $n$', fontsize=12)
#         ax[i, j].set_ylabel('system state $x_{\infty}$', fontsize=12)

fig.text(0.5, 0.0, 'step $n$', ha='center', va='top', fontsize=14)
fig.text(0.0, 0.5, 'system state $x_{\infty}$', ha='right', va='center', rotation='vertical', fontsize=14)

plt.tight_layout()
plt.show()
```

Specifically for smaller values of parameter $r$, one can clearly notice multiple system states $x_{\infty}$

# 3. Plot Bifurcation Diagram for $f(x, r)$

```python
n_params = 1000
n_iterations = 1000
n_last_iterations = 100

x = 0.5
rr = np.linspace(1, 4, n_params)
```

```python
param_ax, state_ax = [], []

for r in rr:
    steps, evolution = calculate_system_evolution(x, r, n_iterations, n_last_iterations)
    
    param_ax.append(r * np.ones(n_last_iterations))
    state_ax.append(np.array(evolution))

param_ax = np.hstack(param_ax)
state_ax = np.hstack(state_ax)
```

```python
plt.figure(figsize=(18, 8))
plt.plot(param_ax, state_ax, ',', alpha=2.)

plt.xlabel('$r$', fontsize=14)
plt.ylabel('system state $x_{\infty}$', fontsize=14)
plt.grid()
plt.show()
```
